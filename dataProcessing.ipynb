### インソールセンサー用ダウンサンプリングコード

```python
# タイムスタンプを削除するコードを追加したほうがいい(タイムスタンプ同期後)
# 正規化と標準化するコードをこっちに移したほうがいい説
# downsampledフォルダを作成するコードを追加した方が良い

rawData_path = './data/20241212test4/InsoleSensor/'

# フォルダ内のすべてのCSVファイルを取得
csv_files = [f for f in os.listdir(rawData_path) if f.endswith('.csv')]

# CSVファイルごとに処理を行う
for csv_file in csv_files:
    # CSVファイルを読み込み
    file_path = os.path.join(rawData_path, csv_file)
    data = pd.read_csv(file_path, header=1)
    
    # タイムスタンプを日付時刻形式に変換
    data["Timestamp"] = pd.to_datetime(data["Timestamp"], unit='s')
    
    # タイムスタンプをインデックスに設定
    data.set_index("Timestamp", inplace=True)
    
    # 100msごとに中央値でダウンサンプリング
    downsampled = data.resample('100L').median() # 50L

    downsampled = downsampled.drop(["Mag_x", "Mag_y", "Mag_z"], axis=1)
    
    # 必要であれば、処理後のデータを保存
    # 下の行のコメントアウトを外して、ファイル名を調整してください。
    downsampled.to_csv(rawData_path + 'downsampled/' + csv_file)

    # 処理内容の確認（例として最初の数行を表示）
    print(f'Processed {csv_file}:')
    print(downsampled.head())
```

### Opti-Track用ダウンサンプリングコード

```python
rawData_path = './data/20241212test4/Opti-track/'

# フォルダ内のすべてのCSVファイルを取得
csv_files = [f for f in os.listdir(rawData_path) if f.endswith('.csv')]

# CSVファイルごとに処理を行う
for csv_file in csv_files:
    # CSVファイルを読み込み
    file_path = os.path.join(rawData_path, csv_file)
    data = pd.read_csv(file_path, header=6)

    data.columns = data.columns.str.strip()

    # タイムスタンプ列を秒単位でDatetime型に変換
    data['Time (Seconds)'] = pd.to_datetime(data['Time (Seconds)'], unit='s')

    # タイムスタンプをインデックスに設定
    data.set_index("Time (Seconds)", inplace=True)
    
    # 100msごとに中央値でダウンサンプリング
    downsampled = data.resample('100L').median()

    # 線形補間で欠損値を埋める
    downsampled = downsampled.interpolate(method='linear')
    
    # 必要であれば、処理後のデータを保存
    # 下の行のコメントアウトを外して、ファイル名を調整してください。
    downsampled.to_csv(rawData_path + 'downsampled/' + csv_file)

    # 処理内容の確認（例として最初の数行を表示）
    print(f'Processed {csv_file}:')
    print(downsampled.head())
```

```python
import numpy as np
import pandas as pd

# data_path = './data/20241212test4/Opti-track/4_timeChecked/,,,,,,,,'

def normalize_skeleton(csv_file, target_center=(0, 0, 0)):
    # target_center: 移動させたい目標の中心座標（x, y, z）
    target_center = np.array(target_center)
    
    # CSVファイルを読み込む
    df = pd.read_csv(data_path + csv_file)
    
    # データを3次元座標の配列に変換
    num_frames = len(df)
    num_joints = len(df.columns) // 3
    coords = np.zeros((num_frames, num_joints, 3))
    
    # データを整形
    for i in range(num_joints):
        coords[:, i, 0] = df.iloc[:, i*3]     # X座標
        coords[:, i, 1] = df.iloc[:, i*3 + 1] # Y座標
        coords[:, i, 2] = df.iloc[:, i*3 + 2] # Z座標
    
    # 正規化されたデータを格納する配列
    normalized_coords = np.zeros_like(coords)
    
    # フレームごとに処理
    for frame in range(num_frames):
        # 1. 現在の中心位置（2番目のポイント）を目標位置に移動
        current_center = coords[frame, 1]
        offset = target_center - current_center
        centered_coords = coords[frame] + offset
        
        # 2. 回転の補正
        hip_vector = centered_coords[13] - centered_coords[17]
        hip_vector[1] = 0  # Y成分を0にして水平面での方向のみを考慮
        
        angle = np.arctan2(hip_vector[0], hip_vector[2])
        
        rotation_matrix = np.array([
            [np.cos(angle),            0, -np.sin(angle)],
            [            0,            1,              0],
            [np.sin(angle),            0,  np.cos(angle)]
        ])
        
        # 回転の中心を目標位置に設定
        for joint in range(num_joints):
            # 目標位置を中心とした座標系に変換
            centered_joint = centered_coords[joint] - target_center
            # 回転を適用
            rotated_joint = rotation_matrix @ centered_joint
            # 目標位置を加算して戻す
            normalized_coords[frame, joint] = rotated_joint + target_center
    
    # データをDataFrameに戻す
    output_data = {}
    for i in range(num_joints):
        output_data[f'X.{i*2+1}'] = normalized_coords[:, i, 0]
        output_data[f'Y.{i*2+1}'] = normalized_coords[:, i, 1]
        output_data[f'Z.{i*2+1}'] = normalized_coords[:, i, 2]
    
    # 結果をCSVファイルとして保存
    output_df = pd.DataFrame(output_data)
    output_df.to_csv(data_path + "5_final/" + csv_file , index=False)

# フォルダ内のすべてのCSVファイルを取得
csv_files = [f for f in os.listdir(data_path) if f.endswith('.csv')]

# CSVファイルごとに処理を行う
for csv_file in csv_files:

    normalize_skeleton(csv_file, target_center=(0, 950, 0))
```
